```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df_ge = pd.read_excel("GE India 2024.xlsx", sheet_name='Final Result')
```

```python
df_ge.info()
```

```python
df_ge['EVM Votes'] = pd.to_numeric(df_ge['EVM Votes'], errors='coerce')
df_ge['Postal Votes'] = pd.to_numeric(df_ge['Postal Votes'], errors='coerce')
df_ge['Total Votes'] = pd.to_numeric(df_ge['Total Votes'], errors='coerce')
df_ge['% of Votes'] = pd.to_numeric(df_ge['% of Votes'].str.replace('%', ''), errors='coerce')
```

```python
df_ge['EVM Votes'].fillna(df_ge['EVM Votes'].median(), inplace=True)
df_ge['Postal Votes'].fillna(0, inplace=True)  # Assuming no postal votes if missing
df_ge['Total Votes'].fillna(df_ge['Total Votes'].median(), inplace=True)
```

```python
victory_margin = df_ge['Victory Margin']
percentage_of_votes = df_ge['% of Votes']

df_ge.drop(columns=['Victory Margin', '% of Votes'], inplace =True )
```
